```python
import pandas as pd
import numpy as np
import sqlite3
import datetime as dt
```

```python
# The datasize is 10 GB so, normal method will be very slower in operation
from sqlalchemy import create_engine

nyc_database = create_engine('sqlite:///nyc_database.db')
```

```python
start = dt.datetime.now()
chunksize = 2000
j = 0
index_start = 1
#VendorID	tpep_pickup_datetime	tpep_dropoff_datetime	passenger_count	trip_distance	RatecodeID	store_and_fwd_flag	PULocationID	DOLocationID	payment_type	fare_amount	extra	mta_tax	tip_amount	tolls_amount	improvement_surcharge	total_amount

for df in pd.read_csv("2018_Yellow_Taxi_Trip_Data.csv",error_bad_lines=False, chunksize=chunksize, iterator=True, encoding='utf-8'):
    df1=df
    df = df.rename(columns={c: c.replace('_', '') for c  in df.columns}) # Remove spaces from columns
    df = df.rename(columns={c: c.replace(' ', '') for c  in df.columns}) # Remove spaces from columns
    print(df.columns)
    #df['pickup_hour'] = [x[11:13] for x in df['tpeppickupdatetime']]
    #df['dropoff_hour'] = [x[11:13] for x in df['tpepdropoffdatetime']]
    #df['tpeppickupdatetime']=pd.to_datetime(df['tpeppickupdatetime'])
    #df['tpepdropoffdatetime']=pd.to_datetime(df['tpepdropoffdatetime'])
    #df.index += index_start
   
    j+=1
    print('{} seconds: completed {} rows'.format((dt.datetime.now() - start).seconds, j*chunksize))
         
    df.to_sql('data', nyc_database, if_exists='append')
    index_start = df.index[-1] + 1
```

```python
df_Ytrip = pd.read_csv("2018_Yellow_Taxi_Trip_Data.csv",low_memory=False,nrows=100000)
df_Ytrip.head()
```

```python
#finding peak hours
df_Ytrip['Days']=pd.DatetimeIndex(df_Ytrip['tpep_pickup_datetime']).day
df_Ytrip['Pickup_time']=pd.DatetimeIndex(df_Ytrip['tpep_pickup_datetime']).hour
df_Ytrip['Drop_time']=pd.DatetimeIndex(df_Ytrip['tpep_dropoff_datetime']).hour
df_Ytrip.head(2)
```

```python
df_Ytrip.groupby('Pickup_time')['Days'].count()
```
